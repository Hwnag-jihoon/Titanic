## 데이터 리스트 목록 
- Previous appearances : 월드컵 출전 횟 수
- Previous titles : 월드컵 우승 횟 수
- Previous finals : 결승 진출 횟 수 
- Previous semifinals : 준 결승 진출 횟 수
- Current FIFA rank : 최신 피파 랭킹 
- Match index : 경기를 구분하는 index
- history with first opponent W-L : 상대 전적(+숫자), 값이 커질 수록 상대 전적에서 앞서 나간 정도 
- history with first opponent goals : 상대 전적 득실차(+숫자), 상대와의 경기 득실차
